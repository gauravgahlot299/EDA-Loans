### Missing Values
Let us look at the missing value percentages in the columns. This is very important as this will help in carrying out Attribute Reduction.

```python
missingValPercent = appData.isnull().sum()/appData.shape[0]*100
print(missingValPercent.sort_values(ascending = False))
```

We are keeping 50% as the threshold and so all those columns which have missing values more than or equal to 50% will be dropped.

```python
appData = appData.loc[:,appData.isnull().sum()/appData.shape[0]*100<50]
print(appData.shape)
appData.head()
```

Selecting the columns which we found with good sense.

```python
appData = appData.loc[:,['SK_ID_CURR',
                        'TARGET',
                        'NAME_CONTRACT_TYPE',
                        'CODE_GENDER',
                        'FLAG_OWN_CAR',
                        'FLAG_OWN_REALTY',
                        'CNT_CHILDREN',
                        'AMT_INCOME_TOTAL',
                        'AMT_CREDIT',
                        'AMT_ANNUITY',
                        'AMT_GOODS_PRICE',
                        'NAME_TYPE_SUITE',
                        'NAME_INCOME_TYPE',
                        'NAME_EDUCATION_TYPE',
                        'NAME_FAMILY_STATUS',
                        'NAME_HOUSING_TYPE',
                        'DAYS_BIRTH',
                        'DAYS_EMPLOYED',
                        'CNT_FAM_MEMBERS',
                        'WEEKDAY_APPR_PROCESS_START',
                        'HOUR_APPR_PROCESS_START',
                        'EXT_SOURCE_2',
                        'EXT_SOURCE_3',
                        ]]
print(appData.shape)
appData.head()
```

```python
missingValPercent = appData.isnull().sum()/appData.shape[0]*100
print(missingValPercent.sort_values(ascending = False))
```

Now we are suggesting few methods for missing value imputations. For this, we will carry few analyses.
We are taking top 5 columns with missing values for this suggestion. Taking 1 at a time.

1. "EXT_SOURCE_3" :- The normalization score column from some source "source 3" is a numerical attribute. Let us see its box plot and then we can decide whether to go ahead with median or mean.

```python
print(appData.EXT_SOURCE_3.mean())
sns.boxplot(y=appData.EXT_SOURCE_3)
```

As can be seen that mean and median are nearly same, we can go ahead with either of them to impute the missing values.

2. "NAME_TYPE_SUITE" :- This column is unordered categorical attribute, therefore we should replace null values with the mode.

3. "EXT_SOURCE_2" :- The normalization score column from some source "source 2" is a numerical attribute. Let us see its box plot and then we can decide whether to go ahead with median or mean.

```python
print(appData.EXT_SOURCE_2.mean())
sns.boxplot(y=appData.EXT_SOURCE_2)
```

As can be seen that mean and median are nearly same, we can go ahead with either of them to impute the missing values.

4. "AMT_GOODS_PRICE" :- It is a numerical attribute. Let us see its box plot and then we can decide whether to go ahead with median or mean.

```python
print(appData.AMT_GOODS_PRICE.mean())
print(appData.AMT_GOODS_PRICE.median())
sns.boxplot(y=appData.AMT_GOODS_PRICE)
```

WE should replace the null values with the median in this case since there are many outliers.

5. "AMT_ANNUITY" :- It is a numerical attribute. Let us see its box plot and then we can decide whether to go ahead with median or mean.

```python
print(appData.AMT_ANNUITY.mean())
print(appData.AMT_ANNUITY.median())
sns.boxplot(y=appData.AMT_ANNUITY)
```

It was very intuitive to say that since annuity amount depend upon the loan amount which is directly related to the good's amount; the metric used for the good's amount should be chosen. Also, the boxplot suggests the same for taking the median for the missing values.

### Checking the datatypes for the attributes

```python
appData.info()
```

We find few attributes with wrong data types. Here is our solution to this :-
1. SK_ID_CURR is unique for every application and it should be unorded categorical; therefore changing it to object.
2. TARGET is actually corresponding to whether the applicant will face difficuly while repaying or not; which shows it has 2 categories. Therefore, changing it to category.
3. CNT_FAM_MEMBERS tells the number of family members the client has. This should not be float but integer. For this, nulls will be replaced with the mode.             

```python
appData.SK_ID_CURR = appData.SK_ID_CURR.astype('object')
appData.TARGET = appData.TARGET.astype('category')
appData.info()
```

```python
toFill = appData['CNT_FAM_MEMBERS'].mode().values[0]
appData['CNT_FAM_MEMBERS'] = appData['CNT_FAM_MEMBERS'].fillna(toFill)
appData['CNT_FAM_MEMBERS'] = appData['CNT_FAM_MEMBERS'].apply(lambda x : int(x))
appData.info()
```

```python
appData.head()
```

We can now see that we have 2 columns pertaining to "Days" which are DAYS_BIRTH and DAYS_EMPLOYED. The values in these are negative. This can be because they have been recorded from the day the application was created. Let us make these positive.

```python
appData.DAYS_BIRTH = abs(appData.DAYS_BIRTH)
appData.DAYS_EMPLOYED = abs(appData.DAYS_EMPLOYED)
appData.head()
```

### Detecting the Outliers
Let's look at the numerical attributes for the outlier detection using box plots.

```python
numCols = appData.select_dtypes(include=['int64','float'])
print(numCols.columns)
```

##### Considering CNT_CHILDREN

```python
sns.boxplot(appData.CNT_CHILDREN)
```

Since there are outliers, let us limit the cap till 2 children (since most value are within 0-2)

```python
appData.CNT_CHILDREN = appData.CNT_CHILDREN[appData.CNT_CHILDREN<=2]
sns.boxplot(appData.CNT_CHILDREN)
```

After removing the outliers, it seems more legit that most of the clients have either 0 or 1 child.

##### Considering AMT_INCOME_TOTAL

```python
sns.boxplot(appData.AMT_INCOME_TOTAL)
```

We have an extreme outlier here. We plan to have some cap on this column of 10,00,00,000

```python
appData.AMT_INCOME_TOTAL = appData.AMT_INCOME_TOTAL[appData.AMT_INCOME_TOTAL<=100000000]
sns.boxplot(appData.AMT_INCOME_TOTAL)
```

We still have outliers. Let us use IQR whiskers for dealing with these.

```python
def standardizeSalary(x):
    if x> upperWhisker:
        return upperWhisker
    else:
        return x    
quantiles = appData.AMT_INCOME_TOTAL.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.AMT_INCOME_TOTAL = appData.AMT_INCOME_TOTAL.apply(lambda x : standardizeSalary(x))
sns.boxplot(appData.AMT_INCOME_TOTAL)
```

We have removed the outliers and this column now seems to be perfect for carrying out the analysis.

##### Considering AMT_CREDIT, AMT_ANNUITY,AMT_GOODS_PRICE

```python
amt = numCols.loc[:,['AMT_CREDIT', 'AMT_ANNUITY','AMT_GOODS_PRICE']]
plt.figure(figsize = (10,10))
sns.boxplot(x="variable", y="value", data=pd.melt(amt))
```

We see there are ouliers in 'AMT_CREDIT', 'AMT_ANNUITY','AMT_GOODS_PRICE'. We are capping the values to 2000000 and so removing the values above it. The rest outliers which remain will be dealt with the IQR whiskers.

```python
appData.AMT_CREDIT = appData.AMT_CREDIT[appData.AMT_CREDIT<2000000]
appData.AMT_GOODS_PRICE = appData.AMT_GOODS_PRICE[appData.AMT_GOODS_PRICE<2000000]

def standardizeAmount(x):
    if x> upperWhisker:
        return upperWhisker
    else:
        return x    
quantiles = appData.AMT_CREDIT.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.AMT_CREDIT = appData.AMT_CREDIT.apply(lambda x : standardizeAmount(x))

quantiles = appData.AMT_ANNUITY.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.AMT_ANNUITY = appData.AMT_ANNUITY.apply(lambda x : standardizeAmount(x))

quantiles = appData.AMT_GOODS_PRICE.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.AMT_GOODS_PRICE = appData.AMT_GOODS_PRICE.apply(lambda x : standardizeAmount(x))

amt = appData.loc[:,['AMT_CREDIT', 'AMT_ANNUITY','AMT_GOODS_PRICE']]
plt.figure(figsize = (10,10))
sns.boxplot(x="variable", y="value", data=pd.melt(amt))
```

After removing the outliers, we are perfectly defined to have amount values in normal scenarios.

##### Considering DAYS_BIRTH, DAYS_EMPLOYED

```python
days = numCols.loc[:,['DAYS_BIRTH', 'DAYS_EMPLOYED']]
plt.figure(figsize = (10,10))
sns.boxplot(x="variable", y="value", data=pd.melt(days))
```

Seeing the value of more than 350000 days in DAYS_EMPLOYED is clearly an outlier. How could a person's employment days be more than his birth days? 
We decide on removing the data for values more than 20000 in DAYS_EMPLOYED and then for the remaining outliers in DAYS_EMPLOYED, we decide on replacing the upper value outliers with the upper whisker value.

```python
appData.DAYS_EMPLOYED = appData.DAYS_EMPLOYED[appData.DAYS_EMPLOYED<20000]
def standardizeDays(x):
    if x> upperWhisker:
        return upperWhisker
    else:
        return x    
quantiles = appData.DAYS_EMPLOYED.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.DAYS_EMPLOYED = appData.DAYS_EMPLOYED.apply(lambda x : standardizeDays(x))
days = appData.loc[:,['DAYS_BIRTH', 'DAYS_EMPLOYED']]
plt.figure(figsize = (10,10))
sns.boxplot(x="variable", y="value", data=pd.melt(days))
```

Now this is legit that employment days are in normal range. This is achieved using outlier detection and treatment.

##### Considering HOUR_APPR_PROCESS_START

```python
sns.boxplot(appData.HOUR_APPR_PROCESS_START)
```

This is weird to know that few people fill up the loan applications in early morning and late nights also (before 5 AM and after 8 PM). 
We decide to replace the outliers in this case with the whisker values. Lower value outliers are replaced with the lower whisker value and upper value outliers are replaced with the upper whisker value.

```python
def standardizeHour(x):
    if x<lowerWhisker:
        return lowerWhisker
    elif x> upperWhisker:
        return upperWhisker
    else:
        return x    
quantiles = appData.HOUR_APPR_PROCESS_START.quantile([0.25,0.75])
IQR = quantiles.values[1] - quantiles.values[0]
lowerWhisker = int( quantiles.values[0]-1.5*IQR )
upperWhisker = int( quantiles.values[1]+1.5*IQR )
appData.HOUR_APPR_PROCESS_START = appData.HOUR_APPR_PROCESS_START.apply(lambda x : standardizeHour(x))
sns.boxplot(appData.HOUR_APPR_PROCESS_START)
```

We capped the loan application hours from 4 in the morning to 8 in the night.

### Conclusions 

1. Age_Bins (which is derived from DAYS_BIRTH) is a strong indicator of knowing whether the client will face the difficult in repaying the loan or not. We have concluded that Young people ought to be more of facing difficulties in repaying; while Middle-aged and Olds not.
2. Other such indicator we found is the Income_Bins (derived from AMT_INCOME_TOTAL). The High income crowd seems to fall less in facing difficulties while repaying the loan amount.
3. Another one is the WEEKDAY_APPR_PROCESS_START. More clients who are at ease of repaying the loans come for loans on Mondays than on Fridays. That is clearcut hint if a client comes on Fridays, he/she might be the one who can face difficulty in repaying back to the bank.
4. Then comes the CNT_CHILDREN. If a client has 2 children there is more probability of him repaying back at ease, in comparison to the ones who bear no child.
5. NAME_FAMILY_STATUS is also important since we noticed that Widows are more liable of facing difficulties in repaying back.
6. Considering NAME_INCOME_TYPE, we see loans can be given to Businessman and Students with low risk rates, since they do not fail in repaying back, while giving loans to people on maternity leaves is risky.
7. DAYS_EMPLOYED says that clients wth less number of employment days are more inclined towards facing difficulties in paying them back.
